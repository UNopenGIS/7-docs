```python
import os
import csv
participation_link = "https://ucla.zoom.us/meeting/register/tJcoc-mvrTovG920aIcgb-64RaKdVWKTb1Ik"
current_markdown_files = []

def process_csv(file):
    with open(file) as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        line_count = 0
        for row in csv_reader:
            if line_count != 0:
                create_markdown_from_csv(row,"en")
            line_count += 1
        print(f'Processed {line_count} lines.')

def get_files_in_folder(folder):
    current_markdown_files = []
    for r, d, f in os.walk(folder):
        for file in f:
            if '.md' in file:
                if file not in current_markdown_files:
                    current_markdown_files.append(os.path.join(r, file))
    return current_markdown_files

def check_list_of_presentation_markdown_files(path):
    return get_files_in_folder(path)

def create_markdown_from_csv(input_csv,language):
    input_date = input_csv[1]

    print('input_date::: '+str(input_date))
    if input_date not in current_markdown_files:
        print('creating file')
        generate_markdown_file(input_csv,language)

def generate_markdown_file(input_csv,language):
    outfile = None
    outfile_path = ""
    
    if language == "en":
        outfile_path = "../docs/events/"+input_csv[1]+".md"
        # outfile = open("../docs/events/"+input_csv[1]+".md", "w")
    else:
        outfile_path = "../i18n/"+language+"/docusarus-plugin-content-docs/current/"+input_csv[1]+".md"
        # outfile = open(, "w")
    if outfile_path is not None:
        append_copy = open(outfile_path, "r")
        original_text = append_copy.read()
        append_copy.close()
        append_copy = open(outfile_path, "w")
        # outfile.write(original_text)
        # write_markdown_header(outfile,input_csv)
        # outfile.write(old_content.read())
        write_presentation_contents(append_copy,input_csv)
   

def write_presentation_contents(outfile,input_csv):
    outfile.write("\n ## Presentation \n"+input_csv[2]+"\n")
    outfile.write("\n View the [recording]("+input_csv[3]+") \n")
    outfile.write(" Download the [slides]("+input_csv[5]+") \n")

def write_markdown_header(outfile,input_csv):
    outfile.write("---\n")
    outfile.write("id: "+input_csv[1]+"\n")
    outfile.write("title: "+input_csv[0]+"\n")
    outfile.write("authors: "+input_csv[2]+"\n")
    outfile.write("sidebar_label: "+input_csv[0]+"\n")
    outfile.write("hide_title: false\n")
    outfile.write("hide_table_of_contents: false\n")
    outfile.write("keywords:\n")
    outfile.write("  - "+input_csv[4]+"\n")
    outfile.write("---\n")

def write_markdown_body(outfile,input_csv):
    outfile.write("\n # "+input_csv[0]+"\n")
    outfile.write("\n ## Date "+input_csv[1]+"\n")
    this_date = input_csv[1]
    this_time = input_csv[2]
    add_padded_zero_for_date(this_date)
    utc_format_the_time(this_time)
    outfile.write("\n Time ["+input_csv[3]+" UTC](https://www.timeanddate.com/worldclock/fixedtime.html?msg="+input_csv[0]+"&iso="+add_padded_zero_for_date(this_date)+"T"+this_time.split(":")[0]+"&p1=1440&ah=1) \n")
    outfile.write("\n ## How to join \n Join via the this [Zoom link]("+participation_link+"). \n")


def add_padded_zero_for_date(this_date):
    this_date = this_date.split("-")
    if len(this_date[1]) == 1:
        this_date[1] = "0"+this_date[1]
    if len(this_date[2]) == 1:
        this_date[2] = "0"+this_date[2]
    this_date = "-".join(this_date)
    return this_date


def utc_format_the_time(this_time):
    this_time = this_time.split(":")
    if len(this_time[0]) == 1:
        this_time[0] = "0"+this_time[0]
    if len(this_time[1]) == 1:
        this_time[1] = "0"+this_time[1]
    this_time = "".join(this_time)
    return this_time

def __main__():
    current_markdown_files_eng = check_list_of_presentation_markdown_files("../docs/events/")
    
    print(current_markdown_files_eng)
    process_csv("../presentations.csv")

if __name__ == "__main__":
    __main__()
```
